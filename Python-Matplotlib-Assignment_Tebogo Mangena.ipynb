```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#1.Size of a figure can be increased by adding the figzise and then specifying the width and height in inches'e.g

plt.figure(figsize=(30,15))
```

```python
#2.Subplots are groups of smaller axes that exist together within a single figure

#Subplots are created by using the command plt.subplot e.g

for i in range (1, 5):
   plt.subplot(2, 2, i)
plt.savefig('subplot.png',dpi=300,bbox_inches='tight') 
plt.show();
```

```python
#Subplots can be navigated by specifically calling a certain subplot and activating it,e.g
plt.subplot(2, 2, 1)
```

```python
#A subplot arrangement can be fixed by using the command plt.tight_layout() 
for i in range (1, 5):
   plt.subplot(2, 2, i)
plt.tight_layout()
plt.savefig('subplot.png',dpi=300,bbox_inches='tight') 
plt.show();
```

```python
#Gridline can be added using the command plt.grid(True)

plt.grid(True)
```

```python
#5 Changing colour,width,style

import numpy as np
import pandas as pd
df=pd.DataFrame({'x_values': range(1,11), 'y_values': np.random.randn(10) })


plt.plot( 'x_values', 'y_values', data=df, color='skyblue')
plt.show()
```

```python
#changing color

plt.plot( 'x_values', 'y_values', data=df, color='green')
plt.show()
```

```python
#Changing width
plt.plot( 'x_values', 'y_values', data=df, linewidth=18)
plt.show
```

```python
#Changing style
# Draw line chart with dashed line
plt.plot( 'x_values', 'y_values', data=df, linestyle='dashed')
plt.show()
```

```python
#6 Changing marker style or color
plt.plot( 'x_values', 'y_values', data=df, marker = 'o', color = 'red')
plt.show()
```

```python
#7 Changing transparency for the lines
# Draw line chart by modifiying transparency of the line
plt.plot( 'x_values', 'y_values', data=df, color='skyblue', alpha=0.3)
plt.show()
```

```python
#8 Adding legend to a plot,the below command is used to add a legend
plt.legend
```

```python
plt.plot( 'x_values', 'y_values', data=df, color='skyblue', alpha=0.3)
plt.legend
plt.show()
```

```python
#9 Changing location of where the legend appears by passing options like upper right, upper left, lower left, lower right, right, center left, center right, lower center, upper center, center to loc argument

#e.g plt.legend(loc = ‘lower right’);

plt.plot( 'x_values', 'y_values', data=df, color='skyblue', alpha=0.3)
plt.legend(loc = 'lower right');
plt.show()
```

```python
#10 Changing the label name on the legend

plt.plot( 'x_values', 'y_values', data=df, color='skyblue', alpha=0.3)
L = plt.legend()
L.get_texts()[0].set_text('line123')
plt.show()
```

```python
#11 Changing size of the legend
plt.plot( 'x_values', 'y_values', data=df, color='skyblue', alpha=0.3)
plt.legend(fontsize = 'x-large');
plt.show()
```

```python
#12 Naming the x and y axis labels and changing the fontsize

plt.plot( 'x_values', 'y_values', data=df, color='skyblue', alpha=0.3)
plt.ylabel('number range', fontsize = 16) #for y label
plt.xlabel('value', fontsize = 16) #for x label
plt.show()
```

```python
#13 Setting the tick marks

plt.plot( 'x_values', 'y_values', data=df, color='skyblue')
plt.yticks([0,2,4,6,8], ['0', '2B','4B', '8B', '10B'])
plt.xticks([0,1,2,3,4], ['0', '1oz', '2oz', '3oz', '4oz'])
plt.show()                           
```

```python
#14 Setting the extent of the axes

plt.plot( 'x_values', 'y_values', data=df, color='skyblue')
plt.ylim(-1.0,1.0) #for y axis
plt.xlim(0, 50) #for x axis
plt.show
```

```python
#15 Changing the scale to logarithmic

plt.plot( 'x_values', 'y_values', data=df, color='skyblue')
plt.yscale('log') #for y axis
plt.xscale('log') #for x axis
plt.show
```

```python
#Bonus Question
#Adding annotations and arrows to a plot

plt.plot( 'x_values', 'y_values', data=df)
plt.annotate('sentosa', xy = (5.0, 3.5), xytext=(4.25, 4.0), arrowprops={'color': 'red'})
plt.savefig('subplot1.png',dpi=300,bbox_inches='tight') 
plt.show
```
